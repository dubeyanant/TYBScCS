```python
import pandas as pd
data = {'Name':pd.Series(['Ram','Sham','Monu','Raj','Sheetal','Pratik','Kunal','Minal','Meetu','Sejal']),'Maths':pd.Series([22,56,78,98,54,76,87,75,88,90]),'Science':pd.Series([99,98,87,76,93,92,82,73,86,71]),'C':pd.Series([99,87,76,65,64,63,69,78,81,44]),'Python':pd.Series([90,82,76,85,72,73,74,65,88,44]),'Java':pd.Series([99,88,87,86,65,73,71,91,88,66])}
df = pd.DataFrame(data)
df
```

```python
from numpy import mean
print('Arithmetic mean:')
print(df.mean())
```

```python
import scipy
print('Geometric mean:')
Maths = scipy.stats.gmean(df.loc[:,'Maths'])
print('Maths\t',Maths)
Science = scipy.stats.gmean(df.loc[:,'Science'])
print('Science\t',Science)
C = scipy.stats.gmean(df.loc[:,'C'])
print('C\t',C)
Python = scipy.stats.gmean(df.loc[:,'Python'])
print('Python\t',Python)
Java = scipy.stats.gmean(df.loc[:,'Java'])
print('Java\t',Java)
```

```python
print('Harmonic mean:')
Maths = scipy.stats.hmean(df.loc[:,'Maths'])
print('Maths\t',Maths)
Science = scipy.stats.hmean(df.loc[:,'Science'])
print('Science\t',Science)
C = scipy.stats.hmean(df.loc[:,'C'])
print('C\t',C)
Python = scipy.stats.hmean(df.loc[:,'Python'])
print('Python\t',Python)
Java = scipy.stats.hmean(df.loc[:,'Java'])
print('Java\t',Java)
```
