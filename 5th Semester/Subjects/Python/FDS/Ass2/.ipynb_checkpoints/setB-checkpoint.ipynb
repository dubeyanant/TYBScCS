```python
#Q2
print(df.mean())
print(df['SepalLengthCm'].value_counts())
print(df['SepalWidthCm'].value_counts())
print(df['PetalLengthCm'].value_counts())
print(df['PetalWidthCm'].value_counts())
print(df['Species'].value_counts())
print('\n')
```

```python
#Q3
print('mean of sepal length ',df['SepalLengthCm'].mean())
print('median of sepal length ',df['SepalLengthCm'].median())
print('\n')
print('mean of sepal width',df['SepalWidthCm'].mean())
print('median of sepal width',df['SepalWidthCm'].median())
print('\n')
print('mean of petal length ',df['PetalLengthCm'].mean())
print('median of petal length ',df['PetalLengthCm'].median())
print('\n')
print('mean of petal width ',df['PetalWidthCm'].mean())
print('median of petal width ',df['PetalWidthCm'].median())
print('\n')
```
